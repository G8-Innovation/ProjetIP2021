# Part 2 : Data Loading

```python
def Load_data(helper_path : str) -> pd.DataFrame:
    """Documentation
    
    Parameters :
        - helper_path : the file path

    Output (if exists):
        - df : My Dataframe cleaned and reindexed

    """
    
    # Data Load with pandas librairy
    df = pd.read_json(helper_path)

    # Drop articles with no content
    df = df[df['art_content'] != '']

    # Reset my dataframe index
    df = df.reset_index(drop = True)
    
    # Returns my clean dataframe
    return df
```

```python
# My file path for the fonction
Helper_path : str = 'C:/Users/Boulanger/Documents/interpromo_2021/df_final.json'

# My DataFrame variable
My_data : pd.DataFrame = Load_data(Helper_path)

# To show my DataFrame
My_data.head(10)
```

# Part 3 : Detection and Translation

```python
def Translate_txt(text : str, language : str) -> str :
    """Documentation
    
    Parameters :
        - text : all summary press articles in the column ["art_content"]
        - language : language selection

    Output (if exists) :
        - text : My text translated

    """
    
    # Get the translation
    try:
        
        # Detection language
        if detect(text) != language :
            
            # Translate my text after detection
            translat = TextBlob(text).translate(to = language).string
            
            # Return my translate text
            return translat
        
        else:
            
            # Return initial text
            return text
    
    except:
        
        # Return initial text
        return text
```

```python
# My initialized column
My_data['art_translat'] : pd.DataFrame = ''

# Translation of each article
for i in trange(len(My_data) , position = 0 , leave = True):
    
    My_data.loc[i, 'art_translat'] = Translate_txt(My_data.loc[i, 'art_content'], 'en')

# To show my DataFrame translated
My_data.head()
```
